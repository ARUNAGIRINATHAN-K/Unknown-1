Annotated Timelines That Speak for Themselves

Your plot is worth nothing if someone has to ask “What’s that spike?”. Annotate it.

```python
import matplotlib.pyplot as plt
import pandas as pd

dates = pd.date_range("2025-01-01", periods=30)
values = [i + (i**1.5 % 5) for i in range(30)]

plt.plot(dates, values, marker='o')
plt.annotate(
    'Major Campaign Launch',
    xy=(dates[10], values[10]),
    xytext=(dates[12], values[10]+2),
    arrowprops=dict(facecolor='red', shrink=0.05)
)
plt.show()
```

2. Ridgeline Plots for Trend Overlap (Not Just for Hipsters)

When multiple distributions overlap over time, ridgeline plots tell the story beautifully.

```python
pip install joypy
```

```python
import joypy
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'value': np.random.randn(2000),
    'month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr'], 2000)
})

joypy.joyplot(df, by='month', column='value', colormap=plt.cm.coolwarm)
plt.show()
```

3. Color Gradients That Tell Magnitude, Not Just Mood

Ditch default blues. Make your color actually mean something.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(10, 10)
plt.imshow(data, cmap='viridis')
plt.colorbar(label='Revenue in $M')
plt.show()
```
