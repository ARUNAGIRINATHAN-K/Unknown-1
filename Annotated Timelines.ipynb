Annotated Timelines That Speak for Themselves

Your plot is worth nothing if someone has to ask “What’s that spike?”. Annotate it.

```python
import matplotlib.pyplot as plt
import pandas as pd

dates = pd.date_range("2025-01-01", periods=30)
values = [i + (i**1.5 % 5) for i in range(30)]

plt.plot(dates, values, marker='o')
plt.annotate(
    'Major Campaign Launch',
    xy=(dates[10], values[10]),
    xytext=(dates[12], values[10]+2),
    arrowprops=dict(facecolor='red', shrink=0.05)
)
plt.show()
```

5. Motion Plots (Animated Stories)

Sometimes the story is in the change, not the snapshot.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation

fig, ax = plt.subplots()
x = np.linspace(0, 2*np.pi, 100)
line, = ax.plot(x, np.sin(x))

def animate(i):
    line.set_ydata(np.sin(x + i/10))
    return line,

ani = animation.FuncAnimation(fig, animate, frames=100, interval=50)
plt.show()
```

6. Slope Charts for Before-and-After Truth Bombs

Bar charts hide the relationship. Slope charts reveal direction instantly.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'category': ['A', 'B', 'C'],
    '2024': [4, 6, 7],
    '2025': [8, 5, 9]
})

for _, row in df.iterrows():
    plt.plot([1, 2], [row['2024'], row['2025']], marker='o')
    plt.text(1, row['2024'], row['category'])
plt.xticks([1, 2], ['2024', '2025'])
plt.show()
```

7. Sankey Diagrams That Explain Complex Flows Instantly

Perfect for showing how inputs become outputs without drowning in arrows.

```python
import plotly.graph_objects as go

fig = go.Figure(go.Sankey(
    node=dict(label=["A", "B", "C", "D"]),
    link=dict(source=[0, 1, 0], target=[2, 3, 3], value=[8, 4, 2])
))
fig.show()
```

8. Highlight-Only-the-Important Bars

Why plot everything the same color when you only care about one bar?

```python
import matplotlib.pyplot as plt

data = [5, 7, 3, 8, 6]
bars = plt.bar(range(len(data)), data)
bars[3].set_color('red')
plt.show()
```

9. Bump Charts for Ranking Over Time

If the plot looks like spaghetti, you’re doing it wrong.

```python
import pandas as pd
import plotly.express as px

df = pd.DataFrame({
    'year': [2022, 2023, 2024, 2025],
    'team': ['A', 'B', 'A', 'B'],
    'rank': [2, 1, 1, 2]
})
px.line(df, x='year', y='rank', color='team', markers=True).show()
```
